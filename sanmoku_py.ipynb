```
import random

list_symble = ["o","x"]
list_action = ["先攻","後攻"]

listA = ["1","2","3","4","5","6","7","8","9"]
listB = [1,2,4,8,16,32,64,128,256]
listC = [7, 56, 448, 73, 146, 292, 273, 84]

listD = [0,0,0,0,0,0,0,0,0]
listE = [0,0,0,0,0,0,0,0,0]

class Person:
  def __init__(self,name,symbol,action):
    self.name = name
    self.symbol = symbol
    self.action = action
    print(self.name + "さんあなたは" + self.symbol + "で" + self.action + "です")


class Line_sum:
  def __init__(self,number,list):
    self.number = number - 1
    self.list = list
    self.sum_list = []
    self.list[self.number] = listB[self.number]

    row1 = self.list[0] + self.list[1] + self.list[2]
    row2 = self.list[3] + self.list[4] + self.list[5]
    row3 = self.list[6] + self.list[7] + self.list[8]

    col1 = self.list[0] + self.list[3] + self.list[6]
    col2 = self.list[1] + self.list[4] + self.list[7]
    col3 = self.list[2] + self.list[5] + self.list[8]

    sla1 = self.list[0] + self.list[4] + self.list[8]
    sla2 = self.list[2] + self.list[4] + self.list[6]

    sum_list = [row1,row2,row3,col1,col2,col3,sla1,sla2]
    self.sum_list = sum_list

  def judg(self):
    if set(self.sum_list) & set(listC):
      return True

    else:
      return False

def call(name,symbol):

    display()

    return int(input(name + "さんどこに" + symbol + "を記入しますか"))

def display():

    print(" " + listA[0] + "| " + listA[1] + "| " + listA[2])
    print("---------")
    print(" " + listA[3] + "| " + listA[4] + "| " + listA[5])
    print("---------")
    print(" " + listA[6] + "| " + listA[7] + "| " + listA[8])


def winner(A,B):
    if A:
        print(Person1.name + "さんの勝ちです！")
    elif B:
        print(Person2.name + "さんの勝ちです！")
    else:
        print("引き分けです！")

def play():
  global Person1,Person2
  player1_symbol = random.choice(list_symble)
  player1_action = random.choice(list_action)

  for s in list_symble:
    if s != player1_symbol:
      player2_symbol = s

  for t in list_action:
    if t != player1_action:
      player2_action = t

  Person1 = Person(input("Please tell me your name:"),player1_symbol,player1_action)
  Person2 = Person(input("Please tell me your name:"),player2_symbol,player2_action)

  input("Please push the Enter key!")
  print("Game Start!")

  turn_count = 0

  while True:
    if not (player1_action == "後攻" and turn_count == 0): #player1が後攻の時1ターン目はスキップする
      number1 = call(Person1.name,Person1.symbol)
      listA[number1 - 1] = Person1.symbol
      turn_count += 1
      sum1 = Line_sum(int(number1),listD)
      judg1 = sum1.judg()

      if judg1 == True or turn_count == 9:
        display()
        break

    number2 = call(Person2.name,Person2.symbol)
    listA[number2 - 1] = Person2.symbol
    turn_count += 1
    sum2 = Line_sum(int(number2),listE)
    judg2 = sum2.judg()

    if judg2 == True or turn_count == 9:
      display()
      break

  winner(judg1,judg2)

play()
```
